```python
import pandas as pd

# Load the dataset
file_path = '/content/Housing.csv'
data = pd.read_csv(file_path)

# Display the first few rows and summary information
print("Initial Data Preview:")
print(data.head())

print("\nData Information:")
data.info()

print("\nBasic Statistical Summary:")
print(data.describe())
```

```python
# Convert categorical columns to numerical values
data['mainroad'] = data['mainroad'].map({'yes': 1, 'no': 0})
data['guestroom'] = data['guestroom'].map({'yes': 1, 'no': 0})
data['basement'] = data['basement'].map({'yes': 1, 'no': 0})
data['hotwaterheating'] = data['hotwaterheating'].map({'yes': 1, 'no': 0})
data['airconditioning'] = data['airconditioning'].map({'yes': 1, 'no': 0})
data['prefarea'] = data['prefarea'].map({'yes': 1, 'no': 0})
data['furnishingstatus'] = data['furnishingstatus'].map({'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0})

# Check for missing values and duplicates
print("\nMissing Values in Each Column:")
print(data.isnull().sum())

print("\nNumber of Duplicate Rows:", data.duplicated().sum())
```

```python
# Show the first few rows after cleaning
print("\nData Preview After Cleaning:")
print(data.head())

# Summary information after cleaning
print("\nData Information After Cleaning:")
data.info()
```

```python
# Calculate mean, median, mode, and other descriptive statistics
mean_values = data.mean()
median_values = data.median()
mode_values = data.mode().iloc[0]  # mode() returns a DataFrame; selecting the first row

print("\nMean Values:")
print(mean_values)

print("\nMedian Values:")
print(median_values)

print("\nMode Values:")
print(mode_values)

print("\nDescriptive Statistics:")
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Housing Prices
plt.figure(figsize=(10, 5))
plt.hist(data['price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Housing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Boxplot for Price by Furnishing Status
plt.figure(figsize=(10, 5))
sns.boxplot(x='furnishingstatus', y='price', data=data, palette='Set2')
plt.title('Price Distribution by Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.show()

# Scatter Plot for Area vs. Price
plt.figure(figsize=(10, 5))
plt.scatter(data['area'], data['price'], color='purple', alpha=0.5)
plt.title('Area vs. Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()

# Heatmap for Correlation
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
